### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```
def RP(df, ind):
  if ind < 4:
    stand_value = df['ë§¤ì¶œì•¡'][0:ind].mean()
    per_val = round((df['ë§¤ì¶œì•¡'][ind]/stand_value - 1)*100, 2)
  else:
    stand_value = df['ë§¤ì¶œì•¡'][ind-4:ind].mean()
    per_val = round((df['ë§¤ì¶œì•¡'][ind]/stand_value - 1)*100, 2)

  if per_val <= -10:
    Relative_Performance = "D"
  elif per_val < -5:
    Relative_Performance = "C"
  elif per_val < 5:
    Relative_Performance = "B"
  elif per_val < 10:
    Relative_Performance = "A"
  else:
    Relative_Performance = "S"   
  return Relative_Performance

RP(df, 4)
```
